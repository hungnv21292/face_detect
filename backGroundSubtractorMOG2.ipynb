```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('Target_and_Obstacle_Recognition.mp4')

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
fgbg = cv2.createBackgroundSubtractorMOG2()

count = 0

while True:
    ret, frame = cap.read()
    
    #print (cap.get(0))
    #print (cap.get(1))
    #print (cap.get(5))
    
    
    if ret == True:
        count += 1
        
        fgmask = fgbg.apply(frame, learningRate=0.01)
        fgmask = cv2.morphologyEx(fgmask, cv2.MORPH_CLOSE, kernel)
        
        # get BackgroundRatio 
        bgRatio = fgbg.getBackgroundRatio()
        
        #print (bgRatio)
        
        # Get Detect Shadown
        detectShadowns = fgbg.getDetectShadows()
        
        if (detectShadowns == False):
            print (detectShadowns)
            
        # Get shadow value
        shadowValue = fgbg.getShadowValue()
        
        #print (shadowValue)
        
        # Get Background image
        bgImage = fgbg.getBackgroundImage()
        
        print (type(bgImage))
        
        print (np.shape(bgImage))
        
        # get the number of last frames that affect the background model
        history = fgbg.getHistory()
        
        #print (history)
        
        # get shadow threshold
        shadowThreshold = fgbg.getShadowThreshold()
        
        print (shadowThreshold)
        
        # get shadow value
        shadowValue = fgbg.getShadowValue()
        print (shadowValue)
        
        # Set shadown threshold
        #fgbg.setShadowValue(155)
        
        #fgbg.setShadowThreshold(0.7)
        
        if count > 50:
            break
        
        
        #cv2.imshow('background', bgImage)        
        cv2.imshow('original', frame)
        cv2.imshow('frame', fgmask)

        if cv2.waitKey(30) & 0xFF == ord('q'):
            break
    
    else:
        break
    
    
                
cap.release()
cv2.destroyAllWindows()
```
